```python
import geopandas as gpd
import folium
import fiona
from owslib.wms import WebMapService
```

## Gewässerlayer einlesen

Das SWISSTLM3D_2026_LV95_LN02.gpkg ist aus Platzgrüden nicht in dem Repository drin.

```python
# swisstopo_wms = WebMapService("http://wms.geo.admin.ch/")
```

```python
pfad = "../Data/Rhein/SWISSTLM3D_2026_LV95_LN02.gpkg"
layer_name = "tlm_gewaesser_fliessgewaesser"
fluesse = gpd.read_file(pfad, layer=layer_name)
fluesse
```

## Rhein herausfiltern

```python
rhein = fluesse[fluesse["name"]== "Rhein | Le Rhin | Rein | Reno"]
rhein
```

## Rhein in Layer speichern

```python
pfad = "../Data/Rhein/"
layer_name = "Rhein"
gdf = gpd.GeoDataFrame(rhein, crs="EPSG:2056")
gdf.to_file(f"{pfad}rhein.gpkg", driver="GPKG", layer=layer_name)
```

## Layer im Geopackage auflisten

```python
pfad = "../Data/Rhein/SWISSTLM3D_2026_LV95_LN02.gpkg"
layers = fiona.listlayers(pfad)
print(layers)
```

## Bodenbedeckungen herausfiltern

```python
pfad = "../Data/Rhein/SWISSTLM3D_2026_LV95_LN02.gpkg"
layer_name = "tlm_bb_bodenbedeckung"
boden = gpd.read_file(pfad, layer=layer_name)
boden
```

## Flüsser herausfiltern

```python
fluesse = boden[boden["objektart"]== "Fliessgewaesser"]
fluesse
```

## Linie vom Rhein mit Gewässerflächen verschneiden

```python
rhein_linie = rhein.geometry.union_all()
rhein_linie
```

```python
rhein_flaechen = fluesse[fluesse.geometry.intersects(rhein_linie)]
rhein_flaechen
```

## Teilfäche herausfiltern

```python
rhein_abs = rhein_flaechen[rhein_flaechen["herkunft_jahr"]== 2025]
rhein_abs
```

## Fläche in Geopackage als neuen Layer schreiben

```python
pfad = "../Data/Rhein/"
layer_name = "Rhein_flaeche_abs"
gdf = gpd.GeoDataFrame(rhein_abs, crs="EPSG:2056")
gdf.to_file(f"{pfad}rhein.gpkg", driver="GPKG", layer=layer_name)
```
