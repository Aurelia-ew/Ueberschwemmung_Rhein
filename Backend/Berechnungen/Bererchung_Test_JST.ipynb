```python
import folium
import geopandas as gpd
import rasterio
import numpy as np
```

```python
pfad = "../Data/Rhein/rhein.gpkg"
layer_name= "Rhein"
rhein= gpd.read_file(pfad, layer=layer_name)

rhein
```

#Fluss einfluss berechnen

```python


with rasterio.open("../Data/Rhein/Hoehenmodell.vrt") as src:

    # Punkte auf Linie erzeugen
    abstand = 10  # Meter

    punkte = []

    linie = rhein.geometry.iloc[0]

    for d in range(0, int(linie.length), abstand):
        punkt = linie.interpolate(d)
        punkte.append((punkt.x, punkt.y))

    # DEM-Höhen sampeln
    hoehen = list(src.sample(punkte))

print(hoehen[:10])
```
