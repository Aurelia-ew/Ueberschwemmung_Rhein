```python
import folium
import rasterio
from folium.raster_layers import ImageOverlay
import numpy as np
```

```python
# Karte
m = folium.Map(location=[47.56240, 7.59661], zoom_start=14, control_scale=True)

# Wasserlevel
levels = [20,15,10,5,4,3,2,1]

for level in levels:

    png = f"../Data/Hoehenmodell/Berechnung/wasserstandplus{level}m.png"
    tiff = f"../Data/Hoehenmodell/Berechnung/veraenderte_hoehen_{level}m_4326.tif"

    # Bounds holen
    with rasterio.open(tiff) as src:
        bounds = [
            [src.bounds.bottom, src.bounds.left],
            [src.bounds.top, src.bounds.right]
        ]

    # Layer erstellen
    fg = folium.FeatureGroup(
        name=f"Wasserstand +{level}m",
        show=False
    ).add_to(m)

    # Overlay hinzufügen
    ImageOverlay(
        image=png,
        bounds=bounds,
        opacity=0.8
    ).add_to(fg)

folium.LayerControl().add_to(m)

m
```
